dummies sert à split les colonnes, par exemple pour marital status, on observe 3 différents type de données.

Avec dummies, 3 nouvelles colonnes seront crées (1 par valeur) afin de répondre par oui ou non a la question.

Par exemple, Situation familiale contient 3 valeurs : Single, maried, Divorced.
    
Dummies créera alors 3 nouvelles colone : single maried et divorced avec soit 1 et 0 pour les valeurs.
    


```python
dummies = pd.get_dummies(
    df
    [
        [
            "Age",
            "Attrition",
            "BusinessTravel",
            "Department",
            "DistanceFromHome",
            "Education",
            "EmployeeID",
            "EmployeeCount",
            "Gender",
            "JobLevel",
            "JobRole",
            "MaritalStatus",
            "MonthlyIncome",
            "NumCompaniesWorked",
            "PercentSalaryHike",
            "StockOptionLevel",
            "TotalWorkingYears",
            "TrainingTimesLastYear",
            "YearsAtCompany",
            "YearsSinceLastPromotion",
            "YearsWithCurrManager",
        ]
    ]
)
dummies
```

```python
pd.crosstab(df.Department, df.Attrition).plot(kind="bar",
                                   figsize=(10, 6),
                                   color=["salmon", "lightblue"])

# Add some communication
plt.title("Out Frequency Per Department")
plt.xlabel("Department")
plt.ylabel("F")
plt.legend(["Yes", "No"])
plt.xticks(rotation=0);
```

```python
pd.crosstab(df.JobRole, df.Attrition).plot(kind="bar",
                                   figsize=(18, 10),
                                   color=["salmon", "lightblue"])

# Add some communication
plt.title("Out Frequency Per JobRole")
plt.xlabel("JobRole")
plt.ylabel("F")
plt.legend(["Yes", "No"])
plt.xticks(rotation=0);
```

```python
for col in dummies.select_dtypes({'int64', 'float64','uint8'}) :
    pd.crosstab(dummies[col], dummies.Attrition).plot(kind="bar",
                                       figsize=(10, 6),
                                       color=["salmon", "lightblue"])

    # Add some communication
    plt.title(f"Out Frequency Per {col}")
    plt.xlabel(col)
    plt.ylabel("F")
    plt.legend(["Yes", "No"])
    plt.xticks(rotation=0);
```
